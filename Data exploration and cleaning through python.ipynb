```python
import pandas as pd
```

```python
df = pd.read_excel('Downloads/Customer Call List.xlsx')
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/._")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('\W','')
df
```

```python
import numpy as np

df = df.replace({np.nan: None, 'N/a': None, 'Na': None})
df
```

```python
#Splitting 'Address' column into 3 columns

df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
df = df.drop(columns="Address")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')

df
```

```python
#If we are giving this data to customer service then we only have to give contacts that they can contact to.
#So now we are removing contacts that have clearly stated that they don't want to be contacted.

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
#So now if there is no phone_number then we can not contact people,so it should be better to drop columns with no contact number

for x in df.index:
    if df.loc[x, "Phone_Number"] == None:
        df.drop(x, inplace=True)
        
df
```

```python
#Resetting index

df.reset_index(drop=True)
df
```
